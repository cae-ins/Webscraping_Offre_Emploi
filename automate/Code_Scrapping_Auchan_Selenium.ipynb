```python
from selenium import webdriver

import time

from selenium.common.exceptions import StaleElementReferenceException

from selenium.webdriver.common.by import By

# Spécifiez le chemin du pilote ChromeDriver
chrome_driver_path = 'C:\\Program Files (x86)\\chromedriver.exe'  # Assurez-vous d'échapper correctement les caractères '\'

# Initialisez le service du navigateur Chrome avec le chemin du pilote
chrome_service = webdriver.chrome.service.Service(chrome_driver_path)

# Initialisez les options du navigateur Chrome
chrome_options = webdriver.ChromeOptions()

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

#from bs4 import BeautifulSoup
#import requests
```

```python
# Initialisez le navigateur Chrome en passant le service et les options
webD = webdriver.Chrome(service=chrome_service, options=chrome_options)

# URL du site Auchan
url = 'https://www.auchan.ci/'
webD.get(url)
```

```python
# Attendre que la page soit complètement chargée (vous pouvez ajuster le délai en fonction de la vitesse de chargement de la page)
webD.implicitly_wait(10)

# Trouver tous les liens des catégories
category_links = webD.find_elements(By.CLASS_NAME, 'jt-stepper-link')
```

```python
# Obtenir tous les liens de groupe d'abord
group_links = [category_link.get_attribute('href') for category_link in category_links]
```

```python
total_products = 0
# Parcourir chaque lien de groupe
#for category_link in category_links:
for group_url in group_links:
    
    try:
        
        # Obtenir le lien du groupe
        #group_url = category_link.get_attribute('href')
    
        # Aller à la page du groupe pour récupérer les informations des produits
        webD.get(group_url)

        # Attendre que la page du groupe soit complètement chargée
        webD.implicitly_wait(10)
    

        # Trouver tous les produits sur la page du groupe
        products = webD.find_elements(By.CLASS_NAME, 'js-jt-product-card')
        
        total_products += len(products)
        
        # Parcourir chaque produit du groupe
        for product in products:
        
            # Obtenir le nom du produit
            product_name = product.find_element(By.CLASS_NAME, 'js-title-line').text

            # Obtenir le prix du produit
            product_price = product.find_element(By.CLASS_NAME, 'js-price-line').text

            # Afficher les caractéristiques et le prix du produit
            print("Nom du produit:", product_name)
            print("Prix du produit:", product_price)
            print("----")
            
        webD.get(url)
     
    except StaleElementReferenceException:
            
        # Si une exception StaleElementReferenceException se produit, réessayez
        print("Stale element reference exception. Retrying...")
        continue

print(total_products)
        
          
```
